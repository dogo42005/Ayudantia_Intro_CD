## Instalación de Anaconda 

Anaconda es una distribución de python y R, que tiene la ventaja de que contiene listas todos las librerías y programas necesarios para hacer ciencia de datos, ahorrándonos tiempo al momento de querer utilizar python ya que nos evita ciertos inconvenientes que podemos tener si instalamos las librerías a mano (Aunque también podemos hacerlo a mano a través de la terminal sin inconvenientes, pero es tedioso)


## Cómo instalamos Anaconda

El sitio web oficial de anaconda es: https://www.anaconda.com/download, donde podrán descargar una versión pertinente a su sistema operativo (Linux, Mac, Windows).


**En caso de contar con un computador de bajos recursos, puede ser mejor realizar la instalación de Miniconda. Es una versión simplificada de Anaconda que se puede utilizar a través de la terminal. Puedes encontrar la información y descargas en https://docs.anaconda.com/miniconda/**


**Si eres usuario Mac, podemos instalar fácilmente Anaconda usando HomeBrew**

- Instalar HomeBrew desde la terminal siguiendo los pasos: https://brew.sh
- Luego de terminada la instalación, ejecutamos lo contenido en el link https://formulae.brew.sh/cask/anaconda (Anaconda) o https://formulae.brew.sh/cask/miniconda (Miniconda)

Usualmente cuando trabajemos en un proyecto de ciencia de datos, las invocaremos haciendo lo siguiente:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
```

Ahora bien, muchas veces nos interesa trabajar con una versión en particular de python porque un proyecto está hecho con una determinada versión de una librería y esa versión es compatible con una versión previa de Python. 

Para solucionar esto, Anaconda nos permite crear instancias de python con sus propias versiones de librerias, las cuales se llaman **ambientes (Enviroments)**

**Les recomiendo que se creen su propio ambiente en Anaconda, lo cual en realidad es muy simple.**

Para crear un ambiente, una vez instalado Anaconda, debemos abrir la terminal (Linux/Mac) o la Anaconda prompt (Windows) y hacer lo siguiente:

* conda create --name **nombre_ambiente** python=3.x -y (Puede ser la versión de python que estimen conveniente. Recomiendo la más actual)

Esto nos va a cargar en la terminal nuestro ambiente. Luego debemos activarlo (Siempre que vayamos a usarlo debemos activarlo)
* conda activate **nombre_ambiente** 

Una vez creado el ambiente, instalamos las librerías que usaremos durante el semestre:
* pip install **libreria** 



Ahora bien, para fortuna de nuestro tiempo, tenemos un archivo requeriments.txt. Este archivo ya contiene todas las librerías que utilizaremos y solo debemos ejecutarlo en nuestro ambiente:
* pip install -r requirements.txt

**No hay problema!, en la carpeta anaconda_config hay un script de Python que crea e instala las librerías por ti! Solo debes primero leer las instrucciones y luego ejecutar el archivo config.py**




##### Adicional: Para eliminar un ambiente **conda remove --name test_env --all**

## Última cosa importante: Como abrir Jupyter Notebooks con ambientes de anaconda


Es simple, una vez creado el ambiente, te vas a la carpeta donde vas a dejar tu archivo .ipynb (Formato Jupyter) y escribes en la terminal

**conda run --name **nombre_ambiente** jupyter notebook**
